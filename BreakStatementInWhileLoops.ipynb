```python
clean_haystack = ["hay","hay","hay","hay",
                  "hay","hay","hay","hay"]

unclean_haystack = ["hay","hay","hay","hay",
                    "hay","needle","hay","hay"]
searching_for = "needle"
```

```python
maximum = len(unclean_haystack)
i = 0
while i < maximum:
    if searching_for == unclean_haystack[i]: # searching for the needle in th unclean_haystack
        print("The %s is at index %i:" %(searching_for, i))
        
        del unclean_haystack[i] # deleting the index number at which the needle was found.
        print('The updated haystack is:', unclean_haystack)
        break
    i = i + 1 # incrementing the index everytime.
else:
    print("\n The needle was not found :(")
```
